```python
import pandas as pd
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('./cow_bite_links_1407_transrac.csv')
df.rename(columns={'Unnamed: 0': ''}, inplace=True)
```

```python
df.head()
```

```python
# Initialize KFold with 10 splits
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Create 10-fold splits
fold = 1
for train_index, valid_index in kf.split(df):
    train_df = df.iloc[train_index]
    valid_df = df.iloc[valid_index]
    
    # Save the train and valid DataFrames as CSV files
    train_df.to_csv(f'train_fold_{fold}.csv', index=False)
    valid_df.to_csv(f'valid_fold_{fold}.csv', index=False)
    
    fold += 1

print("10-fold cross-validation splits have been created and saved as CSV files.")
```
